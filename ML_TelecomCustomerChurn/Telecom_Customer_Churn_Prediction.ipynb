## Data Cleaning

1. Create a copy of base data for manupulation & processing

```python
telco_data = telco_base_data.copy()
```

2. Total Charges should be numeric amount. Let's convert it to numerical data type

```python
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors='coerce')
telco_data.isnull().sum()
```

3. As we can see there are 11 missing values in TotalCharges column. Let's check these records

```python
telco_data.loc[telco_data ['TotalCharges'].isnull() == True]
```

```python
missing = pd.DataFrame((telco_data.isnull().sum())*100/telco_data.shape[0]).reset_index()
plt.figure(figsize=(16,5))
ax = sns.pointplot(x='index',y=0,data=missing)
plt.xticks(rotation =90,fontsize =7)
plt.title("Percentage of Missing values")
plt.ylabel("PERCENTAGE")
plt.show()
```

So, the missing percentage is very less, it is safe to drop those 11 records

4. Missing Value Treatement

Since the % of these records compared to total dataset is very low ie 0.15%, it is safe to ignore them from further processing.

```python
telco_data.dropna(how = 'any', inplace = True)
```

5. Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on...

```python
# Get the max tenure
print(telco_data['tenure'].max())
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

telco_data['tenure_group'] = pd.cut(telco_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
telco_data['tenure_group'].value_counts()
```

6. Remove columns not required for processing



```python
#drop column customerID and tenure
telco_data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
telco_data.head()
```

```python
def uniplot(df,col,title,hue =None):

    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30


    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    #plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright')

    plt.show()
```

## Model Building

```python
df = pd.read_csv('tel_churn.csv')
df.head()
```

```python
# Creating X and Y variables

X = df.drop('Churn', axis = 1)
X
```

```python
y = df['Churn']
y
```

```python
#Train and test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

## Decision Tree Classifier

```python
dtree = DecisionTreeClassifier(max_depth = 6, criterion = 'gini', random_state=100, min_samples_leaf=8)
```

```python
dtree.fit(X_train, y_train)
```

```python
y_pred = dtree.predict(X_test)
```

```python
y_pred
```

```python
dtree.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

```python
print(metrics.confusion_matrix(y_test, y_pred))
```

Since the dependent variable has imbalance the precision, recall and f1-score is minimum

## SMOTEENN

- class used to perform over-sampling using SMOT and cleaning using ENN

```python
sm = SMOTEENN()
X_resampled, y_resampled = sm.fit_resample(X,y)
```

```python
XR_train, XR_test,yr_train,yr_test=train_test_split(X_resampled, y_resampled,test_size=0.2)
```

```python
dtree_resampled = DecisionTreeClassifier(max_depth = 6, criterion = 'gini', random_state=100, min_samples_leaf=8)
```

```python
dtree_resampled.fit(XR_train, yr_train)
```

```python
y_pred_resampled = dtree_resampled.predict(XR_test)
```

```python
y_pred_resampled
```

```python
dtree_resampled.score(XR_test, yr_test)
```

```python
print(classification_report(yr_test, y_pred_resampled, labels=[0,1]))
```

```python
print(metrics.confusion_matrix(yr_test, y_pred_resampled))
```

Better accuracy, Precision, Recall after resampling the dependent variable

## Random Forest Classifier

```python
rf=RandomForestClassifier(n_estimators=100, criterion='gini', random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
rf.score(X_test,y_test)
```

```python
print(classification_report(y_test, y_pred_rf, labels=[0,1]))
```

```python

sm_1 = SMOTEENN()
X_resampled1, y_resampled1 = sm_1.fit_resample(X,y)
```
